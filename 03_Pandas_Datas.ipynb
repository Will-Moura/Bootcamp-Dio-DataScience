```python
import pandas as pd
```

```python
df = pd.read_excel("Aracaju.xlsx")
```

```python
df.dtypes
```

```python
 #Transformar a em tipo datetime.
df["Data"] = pd.to_datetime(df["Data"])
```

```python
#Criando a coluna Receita
df["Receita"] = df["Vendas"].mul(df["Qtde"]) 
df.head()
```

```python
#agrupamento por ano 
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
df["Ano_Vendas"] = df["Data"].dt.year 
df.sample(5)
```

```python
#Extraindo o mes de venda
df["Mês_Vendas"] = (df["Data"].dt.month)
```

```python
#Extraindo o dia de venda 
df["Dia_Vendas"] = (df["Data"].dt.day)
```

```python
#Retornando a data mais antiga. 
df["Data"].min()
```

```python
#Calcular a diferença de dias 
df["Diferença_dias"] = df["Data"] - df["Data"].min() 
 
df.sample(5)
```

```python
#Criando a coluna de trimestre
df["trimestre_venda"] = df["Data"].dt.quarter
df.sample(5)
```

```python
#Filtrar as vendas de 2019 do mês de março 
vendas_marco_2019 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)] 
df.sample()
```
