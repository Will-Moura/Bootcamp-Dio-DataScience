```python
import pandas as pd  
import matplotlib.pyplot  as plt
 
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx") 
 
#Juntar todos os aquivos em um só.
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#agrupamento por ano 
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Transformar a em tipo datetime.
df["Data"] = pd.to_datetime(df["Data"])
```

```python
#Extraindo o dia de venda 
df["Dia_Vendas"] = (df["Data"].dt.day)
```

```python
#Extraindo o mês de venda 
df["Mês_Vendas"] = (df["Data"].dt.month)
```

```python
#Extraindo o ano de venda
df["Ano_Vendas"] = df["Data"].dt.year 
```

```python
#Criando a coluna Receita
df["Receita"] = df["Vendas"].mul(df["Qtde"]) 
```
