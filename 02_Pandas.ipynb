```python
import pandas as pd
```

```python
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
#Juntar todos os aquivos em um só.
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
df.head()
```

```python
df.tail()
```

```python
df.sample(5)
```

```python
df.dtypes
```

```python
#Alterar o tipo do dado,  de inteiro para string. 

df["LojaID"] = df["LojaID"].astype("object")
```

```python
#Verificar se há valores nulos.
df.isnull().sum()
```

```python
#SUBSTITUIR VALORES NULOS PELA MÉDIA DA COLUNA 
df["Vendas"].fillna(df["Vendas"].mean(), inplace = True)
```

```python
df.sample(15)
```

```python
df["Vendas"].mean()
```

```python
#SUBSTITUINDOOS VALORES NULOS POR ZERO
df["Vendas"].fillna(0, inplace=True)
```

```python
#Apagando as linhas com valores nulos.
df.dropna(inplace=True)
```

```python
#Apagando as linhascom valores nulos com base em apenas 1 coluna.
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#Removendo linhas que estejam com valores faltantes em todas as colunas.
df.dropna(how="all", inplace=True)
```

```python
#Criando colunas
df["Receita"] = df["Vendas"].mul(df["Qtde"]) 
df.head()
```

```python
df["Receita"].max()
```

```python
df["Receita"].min()
```

```python
#nlargest top linhas com maiores receitas 

df.nlargest(3,"Receita")
```

```python
#nsamallest 
df.nsmallest(3, "Receita")
```
